```python
api = tweepy.API(auth)
```

```python
def get_tweets(query, maxTweets = 1000):
        #Function to fetch tweets. 
        # empty list to store parsed tweets 
        tweets = [] 
        max_id = -1
        tweetCount = 0
        tweetsPerQry = 100

        while tweetCount < maxTweets:
            try:
                if (max_id <= 0):
                    new_tweets = api.search(q=query +' -filter:retweets', tweet_mode='extended',count=tweetsPerQry,lang="en")
                    
                else:
                    new_tweets = api.search(q=query+' -filter:retweets', tweet_mode='extended', count=tweetsPerQry,
                                                max_id=str(max_id - 1),lang="en")
                    
                if not new_tweets:
                    print("No more tweets found")
                    break

                for tweet in new_tweets:
                    parsed_tweet = {} 
                    parsed_tweet['time & date']=tweet.created_at
                    parsed_tweet['tweets'] = tweet.full_text.replace('\n',' ').encode('utf-8')
                    parsed_tweet['user'] =tweet.user.screen_name.encode('utf-8')
                    parsed_tweet['followers']=tweet.user.followers_count
                    parsed_tweet['likes']=tweet.user.favourites_count
                    
    
                    tweets.append(parsed_tweet) 
                    
                        
                        
                tweetCount += len(new_tweets)
                print("Downloaded {0} tweets".format(tweetCount))
                max_id = new_tweets[-1].id

            except tweepy.TweepError as e:
                # Just exit if any error
                print("Tweepy error : " + str(e))
                break
        
        return pd.DataFrame(tweets)
```
